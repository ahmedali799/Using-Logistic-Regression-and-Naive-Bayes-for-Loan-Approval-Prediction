```python
x = dataset.drop(columns = [' loan_status'])
y = dataset[' loan_status']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=9)
```

### **Model Building**
Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression()
scores = []
names = []
names.append("Logistic Regression")
lr.fit(x_train, y_train)
y_pred_lr = lr.predict(x_test)
score_logreg = accuracy_score(y_test, y_pred_lr)

scores.append(score_logreg)
print("Logistic Regression Accuracy: ", score_logreg)
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(x_train, y_train)
y_pred_nb = nb.predict(x_test)
score_nb = accuracy_score(y_test, y_pred_nb)

scores.append(score_nb)
names.append("Naive Bayes")
print("Naive Bayes Accuracy: ", score_nb)
```

Naive Bayes

```python
plt.bar(names, scores)
plt.xlabel("Algorithms")
plt.ylabel("Scores")
plt.title("Comparison of multiple algorithms")
plt.show()
```
